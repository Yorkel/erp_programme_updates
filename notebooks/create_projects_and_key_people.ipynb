```python
url_rules = {
    "https://www.ucl.ac.uk/education-research-programme": "Education Research Programme",
    "practical-policies-or-bright-ideas": "Education Research Programme",
}
```

```python
def assign_project(text):
    text_lower = text.lower()
    project_hits = set()

    # 1 — Match people → add THEIR project (but don't trigger cross-project on names alone)
    for person, variations in name_variations.items():
        for v in variations:
            if v.lower() in text_lower:
                project_hits.add(project_mapping[person])
                break

    # 2 — Match project title keywords
    for kw, proj in project_title_keywords.items():
        if kw.lower() in text_lower:
            project_hits.add(proj)

    # 3 — Match project URLs
    for url, proj in url_rules.items():
        if url.lower() in text_lower:
            project_hits.add(proj)

    # 4 — Interpret matches based on NUMBER OF UNIQUE PROJECTS
    if len(project_hits) == 0:
        return "Unknown"
    elif len(project_hits) == 1:
        return next(iter(project_hits))    # the only project
    else:
        return "Cross-project"
```

```python
df["project"] = df["text"].apply(assign_project)
```

# Add column for "key people"

```python
def detect_key_people(text):
    text_lower = text.lower()
    persons_found = []

    for person, variations in name_variations.items():
        for v in variations:
            if v.lower() in text_lower:
                persons_found.append(person)
                break  # avoid duplicates
    
    return list(set(persons_found))  # remove duplicates
```

```python
def assign_key_person(text):
    persons = detect_key_people(text)

    if len(persons) == 0:
        return "None"
    elif len(persons) == 1:
        return persons[0]
    else:
        return "Various"
```

```python
df["key_person"] = df["text"].apply(assign_key_person)
```
