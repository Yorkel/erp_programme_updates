```python
url_rules = {
    "https://www.ucl.ac.uk/education-research-programme": "Education Research Programme",
    "practical-policies-or-bright-ideas": "Education Research Programme",
}
```

```python
def assign_project(text):
    text_lower = text.lower()
    matches = set()

    # 1 — Match people
    for person, variations in name_variations.items():
        for v in variations:
            if v.lower() in text_lower:
                matches.add(project_mapping[person])
                break

    # 2 — Match project title keywords
    for kw, proj in project_title_keywords.items():
        if kw.lower() in text_lower:
            matches.add(proj)

    # 3 — Match URLs
    for url, proj in url_rules.items():
        if url.lower() in text_lower:
            matches.add(proj)

    # Interpret matches
    if len(matches) == 0:
        return "Unknown"
    elif len(matches) == 1:
        return list(matches)[0]
    else:
        return "Cross-project"
```

```python
df["project"] = df["text"].apply(assign_project)
```

# Add column for "key people"

```python
def detect_key_people(text):
    text_lower = text.lower()
    persons_found = []

    for person, variations in name_variations.items():
        for v in variations:
            if v.lower() in text_lower:
                persons_found.append(person)
                break  # avoid duplicates
    
    return list(set(persons_found))  # remove duplicates
```

```python
def assign_key_person(text):
    persons = detect_key_people(text)

    if len(persons) == 0:
        return "None"
    elif len(persons) == 1:
        return persons[0]
    else:
        return "Various"
```

```python
df["key_person"] = df["text"].apply(assign_key_person)
```

# Most common words per project 

```python
cv = CountVectorizer(stop_words="english")
```

```python
def get_top_words(texts, n=15):
    X = cv.fit_transform(texts)
    counts = X.sum(axis=0).A1
    vocab = cv.get_feature_names_out()
    freq = list(zip(vocab, counts))
    freq_sorted = sorted(freq, key=lambda x: x[1], reverse=True)
    return freq_sorted[:n]
```

```python
top_words_per_project = {}

for project, group in df.groupby("project"):
    texts = group["text"].astype(str).tolist()
    top_words = get_top_words(texts, n=20)   # choose number of words
    top_words_per_project[project] = top_words
```

```python
for project, words in top_words_per_project.items():
    print(f"\n--- {project} ---")
    for w, c in words:
        print(f"{w}: {c}")
```
